```python
import pandas as pd
```

```python
data =pd.read_excel('Canada.xlsx','Canada by Citizenship',skiprows=range(20), skipfooter=2)
```

```python
data.columns
```

```python
data
```

```python
data = data.set_index(['OdName'])
```

```python
data.dtypes
```

```python
data.dtypes
```

```python
data
```

```python
years = list(map(int,range(1980,2014)))
data.loc['Haiti',years].plot(kind = 'line')
plt.show()
```

```python
data.sort_values(['Total'],ascending = False,axis = 0,inplace = True)
data
```

```python
df_top5 = data.head()
df_top5 = df_top5[years].transpose()
```

```python
df_top5.plot(kind='area')
```

```python
import seaborn as sns
sns.set()
```

```python
count, bin_edges = np.histogram(data[2013])
data[2013].plot(kind='hist',xticks= bin_edges)
```

```python
data_iceland = data.loc['Iceland',years]
data_iceland.plot(kind = 'bar')
```

```python
df_continent = data.groupby('AreaName',axis=0).sum()
```

```python
df_continent = df_continent.drop(['AREA','REG','DEV'],axis=1)
```

```python
df_continent
```

```python
df_continent['Total'].plot(kind='pie')
```

```python
df_japan = data.loc['Japan',years].transpose()
df_japan.plot(kind='box')
```

```python
df_total = pd.DataFrame(data[years].sum(axis = 0))
df_total.reset_index(inplace=True)
df_total.columns = ['year','total']
df_total
```

```python
df_total.plot(kind='scatter',x='year',y='total')
```
