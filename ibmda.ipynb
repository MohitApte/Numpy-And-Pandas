```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from scipy import stats
```

```python
data = pd.read_csv('imports_85.csv',header = None)
data
```

```python
headers = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
data.columns = headers
data
```

```python
data.to_csv('new_data.csv')
```

```python
data['normalized-losses'] = data['normalized-losses'].replace({'?':np.nan})
data 
```

```python
data['normalized-losses'] = data['normalized-losses'].astype('float64')
data['normalized-losses'] = data['normalized-losses'].replace(np.nan,122.000000)
data
```

```python
data['city-mpg'] = 235/data['city-mpg']
data.rename(columns={'city-mpg':'city-L/100km'},inplace = True)
data
```

```python
data.dtypes
```

```python
data['price'] = data['price'].replace({'?':np.nan})
data['price'] = data['price'].astype('float64')
mean = data['price'].mean()
data['price'] = data['price'].replace(np.nan,mean)
data['price'] = data['price'].astype('int64')
data
```

```python
data['normalized-losses'] = data['normalized-losses'].astype('int64')
data
```

```python
print(data.dtypes)
data
```

```python
data['length'] = (data['length'] - data['length'].mean())/(data['length'].std())
data['length']
```

```python
bins = np.linspace(min(data['price']),max(data['price']),4)
group_names = ["Low","Medium","High"]
data["price-binned"] = pd.cut(data['price'],bins,labels=group_names,include_lowest=True)
data
```

```python
data_anova = data[['make','price']]
grouped_anova = data_anova.groupby(['make'])
anova_results = stats.f_oneway(grouped_anova.get_group("honda")['price'], grouped_anova.get_group("jaguar")['price'])
anova_results
```
